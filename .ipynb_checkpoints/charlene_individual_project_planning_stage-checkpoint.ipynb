## (1) Data Description

```r
# loading in the given data for the project
# this is the only loaded file as it is the only one relevant to my research question
players <- read_csv("data/players.csv")
```

```r
# please run the code below to see a summary of statistics
summary(players)

mean_table <- players |>
    summarize(
        "Played Hours" = round(mean(played_hours, na.rm = TRUE), 2),
        "Age" = round(mean(Age, na.rm = TRUE), 2)
    ) |>
    pivot_longer(
        cols = everything(),
        names_to = "Variable",
        values_to = "Mean"
    )
mean_table
```

### Data Source
The data was collected by [a research group at UBC](https://plai.cs.ubc.ca/) from [their self-built MineCraft server](https://plaicraft.ai/).

### Variable Summary
According to the `summary()` output above, the data set `players.csv`contains **196 observations**, where each observation represents a unique player, and **7 variables**.

4 variables fall into the `character` type, including `experience`(the players' self-reported experience of the game), `hashedEmail`(a unique identifier for each player), `name`(the players' in-game name), and `gender`(the players' self-reported gender). 2 of the variables fall into the `numeric` type, which includes `played_hours`(the players' total in-game time on this server) and `Age`(the players' self-reported age). Finally, the variable `subscribe`(whether the players are subscribed to the game-related newsletter) falls into the `logical` type.

### Visible Issues
- The `Age` variable is missing 2 out of 196 values, which will need to be handled with code such as `na.rm = TRUE` before modeling.
- All `character` variables will need to be converted to be used correctly in a predictive model.

### Invisible Issues
- We only have data on 196 players who found and chose to join this research. This does not necessarily generalize data about *all* players. 
- The `experience`, `gender`, and `Age` variables were self-reported, which could lead to potential noise.

# (2) Questions

### Broad question
**Question 1:**
"What player characteristics and behaviours are most predictive of subscribing to a game-related newsletter, and how do these features differ between various player types?"

### Specific Question
"Can a player's `experience`, `played_hours`, `gender`, and `Age` be used to predict their likelihood of **subscribing** (true or false for `subscribe`) to the newsletter?"

### Data Plan
The `players.csv` dataset will be a perfect fit for answering my question, as it contains all the required variables.   

To apply a `k-NN` model, the data must be wrangled in the following ways:
1. Impute the 2 missing `Age` values.
2. Convert `experience` and `gender` to numeric indicator variables.
3. Standardize all numeric predictors so that they are on a common scale.

# (3) Exploratory Data Analysis and Visualization

```r
print("Mean Values for the Quantitative Variables")
mean_table

plot_exp <- players |>
    ggplot(aes(x = experience, fill = subscribe)) +
    geom_bar(position = "fill") +
    labs(title = "Plot 1: Subscription Proportion by Experience", x = "Experience", y = "Proportion", fill = "Subscribed") +
    scale_x_discrete(limits = c("Beginner", "Amateur", "Regular", "Veteran", "Pro")) + 
    scale_fill_discrete(limits = c("TRUE", "FALSE"))
plot_exp

plot_gen <- players |>
    ggplot(aes(x = gender, fill = subscribe)) +
    geom_bar(position = "fill") +
    labs(title = "Plot 2: Subscription Proportion by Gender", x = "Gender", y = "Proportion", fill = "Subscribed") + 
    scale_fill_discrete(limits = c("TRUE", "FALSE"))
plot_gen

plot_hours <- players |>
    ggplot(aes(x = played_hours)) +
    geom_histogram(bins = 20, alpha = 0.7) +
    facet_wrap(~ subscribe) +
    labs(title = "Plot 3: Played Hours Distribution by Subscription", x = "Hours Played", y = "Count") +
    coord_cartesian(xlim = c(0, 40))
plot_hours

plot_age <- players |>
    ggplot(aes(x = Age)) +
    geom_histogram(bins = 15, alpha = 0.7) +
    facet_wrap(~ subscribe) + 
    labs(title = "Plot 4: Age Distribution by Subscription", x = "Age (Years)", y = "Count")
plot_age
```

### Insights
Plot 1 does not show a strong relationship between `subscribe` and `experience`, which hints at `experience` not being a strong predictor.

Plot 2 shows that the subscription portions are very similar for **Female** and **Male** players, the two main self-reported genders, which hints at `gender` not being a strong predictor either.

Plot 3 shows that non-subscribers (`False`) rarely have high-playtime hours; however, subscribers (`True`) are more concentrated near **zero**. `played_hours` could be a better predictor compared to the first two.

Plot 4 shows that subscribers and non-subscribers both peaked around the `Age` 15-20; however, the amount of subscribers at `Age` = 20-25 is far more than that of non-subscribers. Overall, the plots show a weak relationship between `subscribe` and `Age`.

# (4) Methods and Plan

I will use **k-Nearest Neighbors (k-NN)** because my question is a binary classification problem, and k-NN is a strong, non-parametric method for this.

### Assumptions
*k-NN* requires predictors to be on the same scale. Therefore, **standardization** is necessary.

### Limitations
It can be slow on very large datasets and can perform poorly if there are too many predictors.

### Model Selection
I will use **5-fold cross-validation** `vfold_cv()` on the training set to find the best *k*.

### Data Processing Plan
I will split the data into 75% training and 25% testing split with `initial_split()`, making sure that both the training and testing pieces have the same mix of subscribers and non-subscribers as the original data. All data processing will be defined in a `recipe()` that is fit only on the training data and then applied to both the training and testing sets.
